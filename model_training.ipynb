```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# 1. Load the dataset with the engineered feature
df = pd.read_csv('civic_complaints_data_featured.csv')

# 2. Filter data: We can only train on complaints that are actually 'Resolved'
# Drop rows where 'Delay_Days' is NaN (Pending/In-Progress cases)
model_data = df.dropna(subset=['Delay_Days']).copy()

# 3. Select Features (X) and Target (y)
# We ignore Complaint_ID, Filing_Date, Resolution_Date, and Status for the prediction
features = ['Category', 'Area', 'Priority']
X = model_data[features]
y = model_data['Delay_Days']

# 4. Convert text categories into numerical format (One-Hot Encoding)
X_encoded = pd.get_dummies(X, drop_first=True)

# 5. Split the data into Training (80%) and Testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# 6. Initialize and train the Machine Learning Model
print("Training the Random Forest Regressor. Please wait...")
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 7. Make predictions on the test set and evaluate
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)

print("\n--- Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mae:.2f} days")
```

```python
import joblib

# 1. Save the trained model to a file
joblib.dump(model, 'civic_model.pkl')

# 2. Save the exact column names the model was trained on (Very important for deployment)
joblib.dump(list(X_encoded.columns), 'model_columns.pkl')

print("Bhai, Model aur columns successfully '.pkl' files me save ho gaye!")
```
